### Código para formatar a planilha de deputados para o formato *tidy* e salvá-la em csv

```python
import pandas as pd

df = pd.read_csv("clusters_deputados.csv", sep=';')
df
```

```python
df = df.drop(['Unnamed: 0'],axis=1)
```

```python
df.columns
```

```python
formatted_df = pd.melt(df,
                       ['deputado_id', 'deputado_nome', 'deputado_siglaPartido', 'Cluster'],
                       var_name="votação",
                       value_name="freq")
formatted_df = formatted_df.sort_values(by=["deputado_siglaPartido"])
formatted_df.head(10)
```

```python
formatted_df.to_csv("clusters_deputado_tidy.csv", sep=';')
```
